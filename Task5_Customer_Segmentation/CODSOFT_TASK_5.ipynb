```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
```

```python
with open('/content/drive/MyDrive/CodSoft DATASETS/train.wp_source', 'r', encoding='utf-8') as file:
    text = file.read()

print("Total Characters:", len(text))
```

```python
text = text[:200000]

print("New Total Characters:", len(text))
```

```python
print(text[:500])
```

```python
text = text.lower()

print(text[:500])
```

```python
chars = sorted(list(set(text)))

print("Unique Characters:", len(chars))
```

```python
char_to_int = {}
int_to_char = {}

for i, char in enumerate(chars):
    char_to_int[char] = i
    int_to_char[i] = char

encoded_text = []

for char in text:
    encoded_text.append(char_to_int[char])

print(encoded_text[:50])
```

```python
sequence_length = 50

X = []
y = []

for i in range(len(encoded_text) - sequence_length):

    input_sequence = encoded_text[i:i + sequence_length]

    output_character = encoded_text[i + sequence_length]

    X.append(input_sequence)
    y.append(output_character)

print("Total Sequences:", len(X))
```

```python
print("X Samples:", len(X))
print("Y Samples:", len(y))
```

```python
print("Total Sequences:", len(X))
```

```python
X = np.array(X)
y = np.array(y)

print("X Shape:", X.shape)
print("Y Shape:", y.shape)
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))

print("New X Shape:", X.shape)
```

```python
X = X / float(len(chars))

print(X[0])
```

```python
y = to_categorical(y)

print("Y Shape:", y.shape)
```

```python
model = Sequential()

model.add(
    LSTM(
        128,
        input_shape=(X.shape[1], X.shape[2])
    )
)

model.add(Dense(len(chars), activation='softmax'))

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X,
    y,
    epochs=15,
    batch_size=128
)
```

```python
model.save("task5_model.h5")

print("Model Saved Successfully")
```

```python
start_index = np.random.randint(0, len(X)-1)

seed = X[start_index]

generated_text = ""

for i in range(300):

    seed_input = seed.reshape(1, 50, 1)

    prediction = model.predict(seed_input, verbose=0)

    predicted_index = np.argmax(prediction)

    predicted_char = int_to_char[predicted_index]

    generated_text += predicted_char

    seed = np.append(seed[1:], [[predicted_index / len(chars)]], axis=0)

print(generated_text)
```
