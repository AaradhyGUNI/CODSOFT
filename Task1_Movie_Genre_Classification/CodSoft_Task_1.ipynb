```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/drive/MyDrive/CodSoft DATASETS/train_data.txt",
    sep=" ::: ",
    engine="python",
    names=["ID", "TITLE", "GENRE", "DESCRIPTION"]
)
```

```python
X = data["TITLE"] + " " + data["DESCRIPTION"]

y = data["GENRE"]
```

```python
data.dropna(subset=["DESCRIPTION"], inplace=True)
```

```python
tfidf = TfidfVectorizer(
    stop_words="english",
    lowercase=True,
    max_features=100000,
    ngram_range=(1,3),
    min_df=2,
    max_df=0.85,
    sublinear_tf=True
)

X_tfidf = tfidf.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.2,
    random_state=42
)
```

**Naive Bayes**

```python
nb_model = MultinomialNB(alpha=0.3)

nb_model.fit(X_train, y_train)
```

```python
nb_pred = nb_model.predict(X_test)
```

**Accuracy**

```python
nb_accuracy = accuracy_score(y_test, nb_pred)

print("Naive Bayes Accuracy:", nb_accuracy * 100)
```

**Logistic Regression**

```python
lr_model = LogisticRegression(
    max_iter=2000,
    C=4
)

lr_model.fit(X_train, y_train)
```

```python
lr_pred = lr_model.predict(X_test)
```

**Accuracy**

```python
lr_accuracy = accuracy_score(y_test, lr_pred)

print("Logistic Regression Accuracy:", lr_accuracy * 100)
```

**Accuracy of NB and LR**

```python
print("Naive Bayes Accuracy :", round(nb_accuracy * 100, 2), "%")

print("Logistic Regression Accuracy :", round(lr_accuracy * 100, 2), "%")
```

```python
movie ="A poor family struggles to survive in a small village"

movie_tfidf = tfidf.transform([movie])

prediction = lr_model.predict(movie_tfidf)

print("Movie Description :", movie)

print("Predicted Genre :", prediction[0])
```
